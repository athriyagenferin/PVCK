```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pyplot as plt
```

```python

#Membuat list untuk menyimpan url dari beberapa image
urls = ["https://i.pinimg.com/236x/78/d0/48/78d0483f818f30bc2b5f83f2f203e540.jpg",
        "https://i.pinimg.com/564x/8f/6e/64/8f6e6418ad55054bc01a91bc0d9ab062.jpg",
        "https://i.pinimg.com/736x/5a/41/3f/5a413fe88c330bbe194cfff8a7b8f4a8.jpg"]
#baca dan tampilkan image
#loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url) #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #Convert color to RGB
  final_frame = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #show Image
  print("\n")
```

```python

#step 3
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
#step 4
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#horizontal
for y in range(lebar):
  if (y > lebar / 4) and (y < 3 * lebar / 4):
    image_3[int((tinggi)/2),y] = [255,255,255]

cv2_imshow(image_3)
```

```python
#task 4
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#area kotak
kotak_x_start = int(lebar / 4)
kotak_x_end = int(3 * lebar / 4)
kotak_y_start = int(tinggi / 4)
kotak_y_end = int(3 * tinggi / 4)

# Menggambar kotak penuh dengan for loop
for x in range(kotak_y_start, kotak_y_end):
    for y in range(kotak_x_start, kotak_x_end):
        image_3[x, y] = [255, 255, 255]  # Warna kotak (putih)

# hasil
cv2_imshow(image_3)
```

```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io

# Mengambil gambar KTP dari URL
url = "https://i.pinimg.com/564x/b7/30/50/b730507ad166d0b3f1b28b62eed2e82c.jpg"
ktp_image = io.imread(url)
ktp_image = cv.resize(ktp_image, (0, 0), fx=1, fy=1)  # Tidak mengubah ukuran gambar
ktp_image = cv.cvtColor(ktp_image, cv.COLOR_BGR2RGB)  # Mengonversi gambar ke RGB

# Menentukan ukuran dan posisi kotak sensor untuk menutupi Berlaku Hingga
lebar_kotak_sensor = 153  # Lebar kotak sensor
tinggi_kotak_sensor = 19  # Tinggi kotak sensor
start_x = 140  # Koordinat X dari titik awal kotak sensor
start_y = 260  # Koordinat Y dari titik awal kotak sensor

# Menyalin gambar untuk modifikasi
ktp_image_with_sensor = ktp_image.copy()

# Menggambar kotak sensor di area NIK
for y in range(start_y, start_y + tinggi_kotak_sensor):
    for x in range(start_x, start_x + lebar_kotak_sensor):
        if 0 <= x < ktp_image_with_sensor.shape[1] and 0 <= y < ktp_image_with_sensor.shape[0]:
            ktp_image_with_sensor[y, x] = [255, 255, 255]  # Warna Putih
# Menampilkan gambar
cv2_imshow(ktp_image_with_sensor)
```
